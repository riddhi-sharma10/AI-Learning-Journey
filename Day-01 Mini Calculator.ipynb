## MINI CALCULATOR - Python (Rule-Based System)

```python
num1=float(input("Enter the first number:"))
num2=float(input("Enter the second number:"))

operation = input("Enter the operation you want to perform (+,-,*,/,%,**):")

if(operation == '+'):
    print("Result:",num1+num2)
elif(operation == '-'):
    print("Result:",num1-num2)
elif(operation == '*'):
    print("Result:",num1*num2)
elif(operation == '/'):
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Error: Division by zero")
elif(operation == '%'):
    print("Result:",num1%num2)
elif(operation == '**'):
    print("Result:",num1**num2)
else:
    print("Try Again! we have only this much features right now in this calculator.")
```

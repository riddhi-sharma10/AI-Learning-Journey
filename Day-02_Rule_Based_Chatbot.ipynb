## Day 2: Rule-Based Chatbot (AI Fundamentals)

## Project Description
This project implements a simple rule-based chatbot using Python.
The chatbot responds to user queries based on predefined rules,
demonstrating early Artificial Intelligence decision-making systems.

## Objective
- Understand rule-based AI
- Simulate human-like conversation using logic

```python
print("Chatbot: Hi! I am a rule-based chatbot.")
print("Chatbot: Type 'bye' to exit.\n")

while True:
    user_input = input("You: ").lower()

    # Greeting
    if user_input == "hello" or user_input == "hi":
        print("Chatbot: Hello! How can I help you?")

    # Calculator (Day 1 link)
    elif "calculator" in user_input:
        print("Chatbot: I have a mini calculator project built on Day 1.")
        print("Chatbot: It performs operations like +, -, *, /, %, and **.")
        print("Chatbot: It uses rule-based decision-making with if-else logic.")

    # Name
    elif "your name" in user_input:
        print("Chatbot: I am a simple rule-based chatbot.")

    # Status
    elif "how are you" in user_input:
        print("Chatbot: I'm just code, but I'm doing great!")

    # Help
    elif "help" in user_input:
        print("Chatbot: You can say hello, ask about my name,")
        print("Chatbot: ask about the calculator project, or type bye to exit.")

    # Exit
    elif user_input == "bye":
        print("Chatbot: Goodbye! Have a nice day üòä")
        break

    # Default response
    else:
        print("Chatbot: Sorry, I don't understand that yet.")

```
